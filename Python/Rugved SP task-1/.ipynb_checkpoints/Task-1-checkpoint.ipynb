```python
import os
import pandas as pd
df = pd.read_csv("bollywood-1.csv")
count = df['ReleaseTime'].value_counts()
print("No. of movies released in N release time:" ,count[0])
print("No. of movies released in HS release time:", count[1])
print("No. of movies released in FS release time:" ,count[2])
print("No. of movies released in LW release time:" ,count[3])
print("Sorted number of releases in  release time", *count)

cross_tab = pd.crosstab(df['Genre'],df['ReleaseTime'])
print("The maximum number of releases in the Festive season is ", cross_tab['FS'][0])
print(cross_tab)
df['yyyy'] = pd.to_datetime(df['Release Date']).dt.year
date_count = df['yyyy'].value_counts()
print("the year 2014 was the year with the maximum number of movies which is equal to",max(date_count))
df.head()
```

```python
import calendar
df['month']= pd.DatetimeIndex(df['Release Date']).month
df.head()
c= df.loc[(df['Budget']>=30)].groupby("month").size().sort_values(ascending=False).reset_index(name='Count')
c['month'] = c['month'].apply(lambda x: calendar.month_abbr[x])
c.loc[0]
```
